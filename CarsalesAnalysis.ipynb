## A.Data Wrangling Section

### Checking  for null values from the data set.
We use the isnull() method together sum() to solve this problem. 

```python
#Checking for null values in the data set
df.isnull().sum()
```

### Removing Null values from the data set.
This is the method used to handle this: 
  dataframe.dropna(axis=1, how='all', inplace=True)

```python
#Removing null values.
df.dropna(axis=1, how="all", inplace=True)
df.isnull().sum()
```

```python
#Checcking for data types
df.dtypes
```

### Replacing $ with '' and , with '' in our data set.
This will be handy in changing some of our columns data types for proper analysis

```python
# Replacing $ with ''.
df['Price']=df['Price'].str.replace('$', '')
df['Year Resale Value']=df['Year Resale Value'].str.replace('$', '')
df['Retention %']=df['Retention %'].str.replace('%', '')

#Replacing , with ''
df['Price']=df['Price'].str.replace(',', '')
df['Year Resale Value']=df['Year Resale Value'].str.replace(',', '')
df.head(2)
```

### Changing data types for multiple columns.
from object type to interger type.

```python
# Converting multiple colunms to different type, from object type to interger type for better analysis
df[['Year Resale Value', 'Retention %', 'Price']] = df[['Year Resale Value',
                                        'Retention %', 'Price']].apply(pd.to_numeric, 
                                       )
```

# B.Data Analysis Section

   ## 1. Cars Pricing Analysis 
 1. Is there a **correlation** between **car price and yearly resale value**?  
 1. Is there a **correlation** between **car price and engine size**?  
 2. Is there a **correlation** between **car price and fuel efficiency** ?
 1. Is there a **correlation** between **car price and horsepower**?
 2. Is there a **correlation** between **car price and unit sales** ?
 
 
  


```python
#Price and Yearly Resale Value plot 

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Year Resale Value', y='Price')
plt.ylim(0)
plt.xlim(0)
plt.title('Correlation between  yearly resale value and price')
plt.xlabel('Yearly Resale Value')
plt.ylabel('Price')

plt.show()
```

```python
df[['Price','Year Resale Value']].corr()
```

#### Since the Correlation value is **0.9575**.This shows  ***a very strong correlation*** between yearly resale value and price

```python
#Price and Engine size Value plot

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Engine Size', y='Price')
plt.ylim(0)
plt.xlim(0)
plt.title('Correlation between  engine size and price')
plt.xlabel('Engine size')
plt.ylabel('Price')

plt.show()
```

```python
df[['Price','Engine Size']].corr()
```

#### Since the Correlation value is **0.06336**.This shows  **a positive correlation** between engine size and price

```python
#Price and Fuel efficiency plot

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Fuel Efficiency', y='Price')
plt.ylim(0)
plt.xlim(0)
plt.title('Correlation between  fuel effiency and price')
plt.xlabel('Fuel efficiency')
plt.ylabel('Price')

plt.show()
```

```python
df[['Price','Fuel Efficiency']].corr()
```

#### Since the correlation value is **-0.04886.** This shows ***weak correlation*** between fuel efficiency and Price

```python
#Price and Horsepower plot

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Horsepower', y='Price')
plt.ylim(0)
plt.xlim(0)
plt.title('Correlation between horsepower and Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')

plt.show()
```

```python
#Price and Unit of sales plot

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Unit Sales', y='Price')
plt.ylim(0)
plt.xlim(0)
plt.title('Correlation Between Unit Sales and Price')
plt.xlabel('Unit Sales')
plt.ylabel('Price')
plt.show()
```

```python
df[['Unit Sales','Price']].corr()
```

#### Since the correlation value is **-0.30484.** This shows ***weak correlation*** between Unit of Sales and Price
